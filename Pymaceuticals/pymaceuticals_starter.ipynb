```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import traceback

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

merged_data_cleaned = None

# Combine the data into a single DataFrame
try:
    merged_data = pd.merge(study_results,mouse_metadata, on="Mouse ID")
    print("Data merged successfully.")
    merged_data_cleaned = merged_data.drop_duplicates(subset=["Mouse ID", "Drug Regimen"]).head()
except Exception as e:
    print("Error during merging:")
    print(traceback.format_exc())
# Display the data table for preview
merged_data_cleaned

```

```python
# Checking the number of mice.
unique_mice_count = merged_data["Mouse ID"].nunique()
print(f"Number of unique mice: {unique_mice_count}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_data[merged_data.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
merged_data_cleaned = merged_data.drop_duplicates(subset=["Mouse ID", "Timepoint"], keep='first')
merged_data_cleaned.head(5)
```

```python
# Identify duplicates again to ensure none are missed
if not duplicate_mice.empty:
    duplicate_mouse_ids = duplicate_mice["Mouse ID"].unique()
    print(f"Mouse IDs with duplicate entries: {duplicate_mouse_ids}")
   # Remove all entries for the Mouse IDs with duplicates
    merged_data_cleaned = merged_data[~merged_data["Mouse ID"].isin(duplicate_mouse_ids)]
    print(f"Entries for duplicate Mouse IDs removed.")
else:
    # If no duplicates are found, just inform the user
    print(f"No duplicates found.") 
```

```python
# Checking the number of mice in the clean DataFrame.
# Count the number of unique Mouse IDs again
unique_mouse_ids = merged_data_cleaned['Mouse ID'].nunique()
print(f"Number of unique Mouse IDs after cleaning: {unique_mouse_ids}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats =merged_data_cleaned.groupby('Drug Regimen').agg(
   Mean_TumorVolume=('Tumor Volume (mm3)', 'mean'),
    Median_TumorVolume=('Tumor Volume (mm3)', 'median'),
   Variance_TumorVolume=('Tumor Volume (mm3)', 'var'),
    StandardDev_TumorVolume=('Tumor Volume (mm3)', 'std'),
    SEM_TumorVolume=('Tumor Volume (mm3)', 'sem')
)
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

summary_stats = merged_data_cleaned.groupby('Drug Regimen').agg(
    Mean=('Tumor Volume (mm3)', 'mean'),
    Median=('Tumor Volume (mm3)', 'median'),
    Variance=('Tumor Volume (mm3)', 'var'),
    StandardDev=('Tumor Volume (mm3)', 'std'),
    SEM=('Tumor Volume (mm3)', lambda x: st.sem(x))
)
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
promising_treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Start by getting the last (greatest) timepoint for each mouse
final_timepoint = merged_data_cleaned.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_volume = pd.merge(final_timepoint, merged_data_cleaned, on=["Mouse ID", "Timepoint"])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
treatment_volumes = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    treatment_data = final_volume[final_volume["Drug Regimen"] == treatment]["Tumor Volume (mm3)"]
    treatment_volumes.append(treatment_data)
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = treatment_data.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
        
    # add subset 

    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    
    # Determine outliers using upper and lower bounds

    outliers = treatment_data[(treatment_data < lower_bound) | (treatment_data > upper_bound)]
    print(f"{treatment} outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

fig, ax = plt.subplots(figsize=(10, 6))
ax.boxplot(treatment_volumes, labels=treatments,flierprops=dict(markerfacecolor='r', marker='o', markersize=15))
ax.set_title("Tumor Volume Distribution for Each Treatment Group")
ax.set_xlabel("Drug Regimen")
ax.set_ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data = merged_data_cleaned[merged_data_cleaned["Drug Regimen"] == "Capomulin"]
avg_tumor_volume = capomulin_data.groupby("Mouse ID")[["Weight (g)", "Tumor Volume (mm3)"]].mean()

plt.figure(figsize=(6,6))
plt.scatter(avg_tumor_volume["Weight (g)"], avg_tumor_volume["Tumor Volume (mm3)"],marker='o')
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Treatment")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avg_tumor_volume["Weight (g)"], avg_tumor_volume["Tumor Volume (mm3)"])
print(f"Correlation coefficient: {correlation[0]}")

slope, intercept, r_value, p_value, std_err = st.linregress(avg_tumor_volume["Weight (g)"], avg_tumor_volume["Tumor Volume (mm3)"])
regress_values = avg_tumor_volume["Weight (g)"] * slope + intercept

# Plot the linear regression model on top of the scatter plot
plt.figure(figsize=(10,6))
plt.scatter(avg_tumor_volume["Weight (g)"], avg_tumor_volume["Tumor Volume (mm3)"], label='Data')
plt.plot(avg_tumor_volume["Weight (g)"], regress_values, color='red', label='Fit')
plt.title("Average Tumor Volume vs. Mouse Weight for Capomulin Treatment")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.legend()
plt.show()
```
